```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
def func_8a(x):
  return 2 * x

def func_8b(x):
  return 2 * (x-4.5)

x1 = np.arange(-1.5,2)
y1 = func_8a(x1)
plt.plot(x1,y1)

x2 = x1 + 4.5
y2 = func_8b(x2)
plt.plot(x2,y2)



plt.grid(color = '0.8')
plt.show()
```
